```python
import re
```

```python
def clean_title(Movie):
    return re.sub("[^a-zA-Z0-9]","",Movie)
```

```python
df["clean_title"]=df["Movie"].apply(clean_title)
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1, 2))
t = vectorizer.fit_transform(df["clean_title"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
Movie="War"
movie=clean_title(Movie)
query=vectorizer.transform([movie])
similarity=cosine_similarity(query,t).flatten()
indices=np.argpartition(similarity,-10)[-10:]
results=df.iloc[indices]
```
